```python
#сортировка
df=pd.read_excel('atom.xlsx')
df = df.dropna()
#Сортировка данных
sorted_df = df.sort_values(by='Атомные электростанции России', ascending=True) 
sorted_df
```

```python
#добавление столбца
sred = df['Мощность (МВт)'].mean()
df1 = df.assign(Отклонение_отсреднего=df['Мощность (МВт)'] - sred)
df1
```

```python
#Выбор фрагмента таблицы по номерам строк и столбцов.
filtered_df = df.iloc[[0, 1, 2, 3], [0, 1, 2]] # Выберите строки с индексами 0, 1, 2, 3,  а также столбцы с индексами 0, 1, 2
filtered_df
```

```python
#Расчеты статистических показателей по сгруппированным данным.
total_sales_by_category = df.groupby('Субъект РФ (область)')['Мощность (МВт)'].sum()
print(total_sales_by_category)
```

```python
#Построение гистограммы по данным.
plt.bar(df['Атомные электростанции России'], df['Мощность (МВт)'])
plt.xticks(rotation=90)
plt.xlabel('Атомные электростанции России')
plt.ylabel('Мощность (МВт)')
plt.title('Гистограмма мощности АЭС')

plt.show()
```

```python
#Построение круговой диаграммы по данным.
plt.pie(df['Мощность (МВт)'], labels=df['Атомные электростанции России'])
plt.title('Круговая диаграмма мощности АЭС')
plt.show()
```

```python
#Построение интерактивного графика по данным.
fig = px.bar(df, x='Атомные электростанции России', y='Мощность (МВт)')
fig.show()
     
```
